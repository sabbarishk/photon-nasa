```
# cell 1
import xarray as xr
url = "https://example.com/path/to/dataset.nc"
ds = xr.open_dataset(url)

# cell 2
var = ds.get('surface_reflectance') if 'surface_reflectance' in ds else None
import matplotlib.pyplot as plt
if var is not None:
    var_mean = var.mean(dim='time', skipna=True)
    var_mean.plot()
    plt.title('Mean Surface Reflectance Map')
    plt.show()
```
